```python
import pandas as pd
import requests
from tqdm import tqdm

tqdm = tqdm.pandas()
```

```python
df = pd.read_csv('hospital.csv', dtype=str)
```

```python
def download_file(url):
    filename = url.split('/')[-1]
    r = requests.get(url, stream=True).content
    
    with open('.\\input_files\\' + filename, 'wb') as f:
        f.write(r)
```

```python
df['stdchg_file_url'].progress_apply(download_file)
```
