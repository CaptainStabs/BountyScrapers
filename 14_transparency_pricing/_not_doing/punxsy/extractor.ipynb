```python
import pandas as pd
```

```python
file = 'Punxsy Price Transparency Final (1).csv'
df = pd.read_csv(file)
```

```python
df = df.rename(columns={
    'Charge Code': 'procedure_code',
    'Description': 'description',
    'Revenue Code': 'rev_code',
    'CPT/HCPCS': 'code'
})

df.columns = df.columns.str.strip()
```

```python
df.drop(columns=['Unnamed: 18'], inplace=True)
```

```python
cols = df.columns.tolist()
cash_index = cols.index('Cash Price')
id_vars = cols[:cash_index]
id_vars.remove('Hospital Price')

value_vars = cols[cash_index:]
value_vars.append('Hospital Price')

df = pd.melt(df, id_vars=id_vars, value_vars=value_vars, var_name='payer_orig', value_name= 'rate')
```

```python
df['rate'] = df['rate'].str.replace("$", "").str.replace(",", "")
```
